## Univariate Linear Regression

In the correlation we found that Total Income is most correlated with Total Expenditure after Total Household Income.
So We predict Total Income and Total Expenditure in Univariate Form.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = data[['Total Income']]
y = data['Total Expenditure']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
model.fit(x_train,y_train)
```

```python
print('Intercept of Total Expenditure is :' , model.intercept_.round(2))
```

```python
print("Coefficient of Total Income is :" , model.coef_.round(2))
```

```python
y_test_predicted = model.predict(x_test)
```

```python
data['Total Income'].max()
```

```python
data['Total Income'].min()
```

```python
data['Total Expenditure'].max()
```

```python
data['Total Expenditure'].min()
```

```python
plt.scatter(x_test,y_test,label='Tested Data')
plt.plot(x_test,y_test_predicted,label='Prediction',color='g')
plt.xlabel('Income')
plt.ylabel("Expenditure")
plt.xlim((12000,5000000))
plt.ylim((18000,5300000))
plt.title('Tested Data vs Predicted Data')
plt.legend(loc='upper left')
plt.show()
```

In the above graph, blue scattered chart is the data which we get by train_test_split model and the blue straight line is our predicted model.

```python
#residuals
```

```python
residuals = y_test - y_test_predicted
```

```python
plt.scatter(x_test,residuals,label='Residual Data')
plt.hlines(y=0,xmin=x_test.min(),xmax=x_test.max(),linestyle='--',color='k',label='Normal')
plt.xlabel('Income')
plt.ylabel('Expenditure')
plt.title('Residuals of Tested and Predicted')
plt.legend(loc='lower right')
plt.xlim((12000,5000000))
plt.show()
```

In the above graph, We selected the data of errors in the test and predicted model and this horizontal line is centered point which shows that our prediction errorness is nearly 0.

```python
from sklearn.metrics import mean_squared_error
```

```python
print("Standard Deviation of y_test is :" , y_test.std().round(2))
```

```python
print("Standard Deviation of y_test_predicted is :" , y_test_predicted.std().round(2))
```

```python
print("Mean Squared Error of univariate is :" , mean_squared_error(y_test,y_test_predicted).round(2))
```

```python
print("R-Squared value is :" , model.score(x_test,y_test).round(2))
```
